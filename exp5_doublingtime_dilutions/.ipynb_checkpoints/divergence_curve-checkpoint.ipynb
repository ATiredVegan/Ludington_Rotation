# Divergence from first OD over time 
Plates were diluted three times 2:1 at doubling time (90 minutes). Waited for OD of 0.4 ish. 
We will take the OD at the last time point on each plate, average it over replicates. Should also normalize to initial OD.

```python
dilutions=pd.read_excel("exp5_data.xlsx")
plate=pd.read_excel("exp5_plate.xlsx",index_col=0)
order=['Dilution']
for well in plate.itertuples():
    for entry in well[1:]:
        order.append(entry)
dilutions.columns=order
dilutions

       
```

```python
samples=['GFP','MC']
for sample in samples:
    dilutions['NormAve '+sample]=(dilutions[sample+' 1x'].mean(axis=1)-dilutions['1x Media'].mean(axis=1))
    dilutions['SD '+sample]=dilutions[sample+' 1x'].std(axis=1)
```
